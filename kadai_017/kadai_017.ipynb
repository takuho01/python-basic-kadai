```python
# kadai 017

# class : human (name , age, )
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def check_adult(self):
        if self.age >= 20:
            print("大人です")
        else:
            print("大人じゃないです")

Human1 = Human("bob", 20)
Human2 = Human("michael", 18)
Human3 = Human("jun", 25)


list_human = [ Human1, Human2, Human3 ]

for person in list_human:
    person.check_adult()




```
